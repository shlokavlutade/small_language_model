### Dataset Preparation

```python
from datasets import load_dataset
```

```python
train_dataset = load_dataset("roneneldan/TinyStories", split="train")
validation_dataset = load_dataset("roneneldan/TinyStories", split="validation")
```

```python
print(train_dataset[0:3])
# print(validation_dataset[0])
```

```python
def save_txt(dataset, filename):
    with open(filename, "w", encoding="utf-8") as f:
        for story in dataset['text']:
            story = story.strip().replace("\n", " ")
            f.write(story + "\n<|endoftext|>\n")
```

```python
save_txt(train_dataset , 'train.txt')
save_txt(validation_dataset , 'validation.txt')
```

```python
import numpy as np
import tiktoken

def load_text(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        return f.read()
    
def tokenize_text(text, tokenizer):
    return tokenizer.encode(text, allowed_special= {"<|endoftext|>"})

def save_bin(tokens, filename):
    np.array(tokens, dtype= np.uint16).tofile(filename)

def create_bin_files(train_txt, val_txt, train_bin, val_bin):
    
    tokenizer = tiktoken.get_encoding("gpt2")
    train_txt = load_text('train.txt')
    val_txt = load_text('validation.txt')

    train_tokens = tokenize_text(train_txt,tokenizer)
    val_tokens = tokenize_text(val_txt, tokenizer)

    print(f"Training tokens: {len(train_tokens):,}")
    print(f"Validation tokens: {len(val_tokens):,}")

    save_bin(train_tokens, train_bin)
    save_bin(val_tokens, val_bin)

    print("Created binary files")
```

```python
create_bin_files(train_txt="train.txt", val_txt="validation.txt", train_bin="train.bin",
    val_bin="val.bin")
```
